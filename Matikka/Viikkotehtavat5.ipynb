```python
import numpy as np
import matplotlib.pyplot as plt
```

Tehtävä 1

```python
# Funktio ja sen derivaatta
def f(x):
    return x**4 + 5*x**3 - 7*x

def df(x):
    return 4*x**3 + 15*x**2 - 7

# Aloituspiste ja oppimisnopeus
x = 0.0
alpha = 0.01
path_x = np.array([x])

# Gradient descent -iterointi
for k in range(1000):
    x = x - alpha * df(x)
    path_x = np.append(path_x, x)

# Piirretään polku
path_y = f(path_x)

# Tarkastellaan arvoja
print(f"Löydetty minimi: x ≈ {x:.6f}")
print(f"f(x) ≈ {f(x):.6f}")

# Piirretään polku ja funktio
xx = np.linspace(-5, 3, 1000)
yy = f(xx)

plt.plot(path_x, path_y, '.', label='Gradient descent polku')
plt.plot(xx, yy, label='f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.title('Gradient Descent -polku')
plt.show()
```

Tehtävä 2

```python
def f(x, y):
    return 3*x**2 + 3*x*y + 2*y**2 + 3*x - 5*y

def df(x,y):
    dx = 3*2*x + 3*y +3
    dy = 3*x + 2*2*y - 5
    return dx, dy

def calc_g_descent(x, y, alpha):
    for k in range(100):
        dx, dy = df(x, y)
        x = x - alpha*dx
        y = y - alpha*dy

    print("x = ", x, "     y = ", y)

# Testi 1: Aloituspiste (0, 5), alpha = 0.02
calc_g_descent(0, 5, 0.02)

# Testi 2: Aloituspiste (-5, -5), alpha = 0.01
calc_g_descent(-5, -5, 0.01)

# Testi 3: Aloituspiste (10, -10), alpha = 0.03
calc_g_descent(10, -10, 0.03)

# Lopuksi testataan vielä ylhäällä olevista parhaita parametreja ja lisätään hieman oppimisnopeutta
calc_g_descent(0, 5, 0.05)
```
